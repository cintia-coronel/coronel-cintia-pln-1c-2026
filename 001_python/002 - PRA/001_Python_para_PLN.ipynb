## Ejercicio Práctico 2: Contador de Frecuencias

> **Consigna:** A partir del texto dado, escribí un programa que lo convierta a minúsculas, lo divida en palabras y cuente cuántas veces aparece cada una.
>
> El resultado debe ser un diccionario donde las claves sean las palabras y los valores sean sus frecuencias.

**Pasos sugeridos:**

1.  Convertí el texto a minúsculas con `.lower()`.
2.  Dividilo en palabras con `.split()`.
3.  Recorrí la lista con un `for`.
4.  Para cada palabra, actualizá su contador en el diccionario.

**Preguntas para trabajar con IA:**

- *"¿Para qué sirve `dict.get(clave, 0)` al contar frecuencias?"*
- *"¿Cómo ordenar un diccionario por sus valores de mayor a menor?"*
- *"¿Hay una forma más concisa de hacer un conteo de frecuencias en Python?"*

---

```python
# Ejercicio: Contador de palabras
texto = "Python es un lenguaje de programación ampliamente utilizado en el procesamiento del lenguaje natural"

# Tu código acá:
# 1. Convertir texto a minúsculas
texto_lower = texto.lower()
# 2. Dividir en palabras
palabras = texto_lower.split()
# 3. Contar frecuencia de cada palabra
frecuencias = {}
for palabra in palabras:
    if palabra in frecuencias:
        frecuencias[palabra] = frecuencias[palabra] + 1
    else:
        frecuencias[palabra] = 1

print(frecuencias)

print("¡Intenta resolver este ejercicio!")
```
